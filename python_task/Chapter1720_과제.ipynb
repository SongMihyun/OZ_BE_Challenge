#### 리팩토링 과제

Day4에서 만든 "붕어빵 자동 판매 시스템"은 절차적 코드로 구현되어있습니다.

Day5에서 배운 함수, 클래스 개념을 적용시켜 객체지향적으로 리팩토링을 진행해주세요


```python
# 리팩토링 과제 힌트
# 구조는 아래와 같습니다.
class BungeoppangShop:
    def __init__(self):
        self.종류 = ['팥붕어빵', '슈크림붕어빵', '초코붕어빵']
        self.재고 = [10,8,5]
        self.가격 = [1000,1200,1500]
        self.현황판 = dict(zip(self.종류,self.재고))
        self.판매수량=[0,0,0]
        self.상품별매출=[0,0,0]
        self.일일총매출 =0
        self.admin='n'
        self.order_stop='n'
        self.현황판_new=self.현황판

    def order_page(self, bread_type, bread_count):
        while self.order_stop != 'y' : 
            
            if self.order_stop=='y': 
                continue
            else:
                try:
                    self.order_종류 = int(input("1.팥  2.슈크림  3.초코 / 숫자로 선택하세요"))  
                    if self.order_종류 > len(self.종류) or self.order_종류 <= 0 :
                        print('없는 주문번호 입니다.') 
                    else:
                        self.order_수량 = int(input(f'{self.order_종류} 몇개?'))
                        self.판매수량[self.order_종류-1] += self.order_수량
                        print(f'{self.종류[self.order_종류-1]} : {self.order_수량} 개를 주문하셨습니다.')
                except ValueError: 
                    print("숫자로 주문 부탁드립니다.")

                if  self.재고[self.order_종류-1] >= self.order_수량:
                    self.재고[self.order_종류 -1] -= self.order_수량
                    self.현황판_new = dict(zip(self.종류,self.재고))
                    self.order_count = dict(zip(self.종류,self.판매수량))
                    
                else:
                    print(f'현재 {self.종류[self.order_종류-1]}의 재고가 {self.재고[self.order_종류-1]}개 입니다. 넘지 않게 다시 주문해주세요')
                self.order_stop=input("주문을 멈추시겠습니까? y/n")
                
    def order_page(self):
        order_stop='n'
        while order_stop != 'y' : 
            
            if order_stop=='y': 
                continue
            else:
                try:
                    order_종류 = int(input("1.팥  2.슈크림  3.초코 / 숫자로 선택하세요"))  
                    if order_종류 > len(self.종류) or order_종류 <= 0 :
                        print('없는 주문번호 입니다.') 
                    else:
                        order_수량 = int(input(f'{order_종류} 몇개?'))
                        self.판매수량[order_종류-1] += order_수량
                        print(f'{self.종류[order_종류-1]} : {order_수량} 개를 주문하셨습니다.')
                except ValueError: 
                    print("숫자로 주문 부탁드립니다.")

                if  self.재고[order_종류-1] >= order_수량:
                    self.재고[order_종류 -1] -= order_수량
                    현황판_new = dict(zip(self.종류,self.재고))
                    order_count = dict(zip(self.종류,self.판매수량))
                    
                else:
                    print(f'현재 {self.종류[order_종류-1]}의 재고가 {self.재고[order_종류-1]}개 입니다. 넘지 않게 다시 주문해주세요')
                order_stop=input("주문을 멈추시겠습니까? y/n")
                
            
        print(f'총 주문수량은 {order_count}')
        print(f'남은 재고 : {현황판_new}')

    def add_page(self):
        admin = input('관리자모드를 진행하시겠습니까? y/n') # 관리자모드 진입여부 질문
        현황판_new = self.현황판
        while admin =='y':   #관리자 모드 일때
            추가품목명 = input("추가할품목은? / 1.팥붕어빵 , 2.슈크림붕어빵 / 3.초코붕어빵 /종료")
            if 추가품목명 == '종료':
                admin = 'n' 
                print('관리자모드가 종료되었습니다.') 
                break 
            else:  
                admin='y'
                try:    #추가할 품목을 숫자로 받았을때
                    추가품목명_index=int(추가품목명)    #받은 정보를 숫자로 형변환
                    if 0 < 추가품목명_index <= len(self.종류):   # 품목이 있는지 확인(종류의 갯수로 확인)
                        추가수량 = int(input("추가할 수량은?")) #추가수량 입력받아 형변환
                        품목명=self.종류[추가품목명_index-1] # 받은 숫자로 해당 품목명을 호출
                        현황판_new[품목명]+=추가수량
                        print(f'{품목명} : {추가수량} 이 추가되었습니다.')
                    else:
                        print('입력하신 상품이 존재하지 않습니다. 숫자를 다시 확인해 주세요')
                except ValueError:
                    if 추가품목명 in 현황판_new:    #현황판 키값에 추가품목명이 있는지 확인
                        try :
                            추가수량 = int(input("추가할 수량은?")) #추가할 수량 입력받아 int로 형변환    
                            현황판_new[추가품목명]+=추가수량
                            print(f'{추가품목명} : {추가수량} 이 추가되었습니다.')
                        except ValueError:
                            print('수량을 숫자로 정확히 입력해 주세요~!')
                    else:
                        print('입력하신 상품을 찾을수 없습니다. 품목명을 정확히 입력해주세요.')
                        
                    
                print(f'현재 재고 : {현황판_new}')               
            
    def 단가(self):
        print("------ 상품별 단가 ------")
        for i in range(len(self.종류)):
            print(f'{self.종류[i]}: {self.가격[i]}')
            
    def 일일판매량(self):
        print("------ 상품별 판매량 ------")
        for i in range(len(self.종류)):
            print(f'{self.종류[i]}:{self.판매수량[i]}개')
            
    def 일일매출(self):
        일일총매출 = 0
        print("---------------------")
        print("------ 일일 매출 ------")
        print("---------------------")
        
        for i in range(self.len(self.종류)):
            self.상품별매출[i] = self.가격[i] * self.판매수량[i]
            일일총매출 += self.상품별매출[i]    
        print(f'일일총매출 : {일일총매출}')

    def mode_cho(self):
        mode_input = input("1. 주문 ") #1번누르면 주문 / 9999번누르면  관리자모드
        try :
            mode_int=int(mode_input)
            if mode_int == 9999:    #관리자모드
                print("관리자모드입니다.")
                self.admin_page()
            else:   # 주문모드
                print("주문모드입니다.") 
                self.order_page()       
        except:
            print("숫자로 입력해주세요")
            
    def admin_page(self):
        mode_input = input("1.매출 / 2.재고관리")
        try :
            mode_int=int(mode_input)
            if mode_int == 1:    
                self.단가()
                self.일일판매량()
                self.일일매출()
            else:   # 재고관리
                print('재고관리 모드')
                self.add_page() 
        except:
            print("숫자로 입력해주세요")
붕어빵=BungeoppangShop()
붕어빵.mode_cho()
```
